```python
dependent = df['outcome']
```

```python
from sklearn import tree
decision = tree.DecisionTreeClassifier()
```

```python
decision.fit(independent, dependent)
```

```python
decision.score(independent, dependent)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig = fig = plt.figure(figsize=(100,50))
print(tree.plot_tree((decision),filled=1))
```
